### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.

```python
#JLT HW6-APIs

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import csv 
import openweathermapy.core as owm
from config import api_key
from citipy import citipy
import datetime
import gmaps

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#create a url
url = "https://api.openweathermap.org/data/2.5/weather?q="

#create lists to hold data
lat = []
lng = []
max_temp = []
humidity = []
wind_speed = []
cloudiness = []
date = []
country = []
available_cities = []

#create counters for set number
counter = 0
set_number = 1


print("Begining Data Retrieval")
print("-----------------------")

for index, city in enumerate (cities):

    query_url = url + city + "&units=Imperial&APPID=" + api_key
    
    # assemble url and make API request
    response = requests.get(query_url).json()
    #print(response)
    
    counter = counter + 1
    
    if counter == 49:
        time.sleep(3)
        set_number = set_number + 1
        counter = 0
    
    try:

        print(f"Processing Record {index} of Set {set_number} | {cities[index]}")
        print(f"{url}units=Imperial&APPID={api_key}&q={cities[index]}")
        
        lng.append(response["coord"]["lon"])
        lat.append(response["coord"]["lat"])
        max_temp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        wind_speed.append(response["wind"]["speed"])
        cloudiness.append(response["clouds"]["all"])
        date.append(response["dt"])
        country.append(response["sys"]["country"])
        available_cities.append(response['name'])
    
    #exception handling 
    except:
        print("City not found. Skipping...") 
        
        continue

# get the date ran the data for headers in plots
plotheaderdate = datetime.datetime.now()
#print(plotheaderdate.strftime("%x"))        
        
print("-----------------------")
print("Data Retrieval Complete")
print("-----------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#put lists into dataframe and save to csv
cities_df = pd.DataFrame({"City": available_cities,
                         "Cloudiness": cloudiness,
                         "Country": country,
                         "Date": date,
                         "Humidity": humidity,
                         "Lat": lat,
                         "Lng": lng,
                         "Max Temp": max_temp,
                         "Wind Speed": wind_speed})
# cities_df
cities_df.to_csv(output_data_file, encoding="utf-8", index=False)
cities_df.count()
```

#### Latitude vs. Temperature Plot

```python
# Obtain the x and y coordinates 
plt.ylim(min(max_temp),max(max_temp))
plt.xlim(min(lat),max(lat))

# Build the scatter plot
x_axis1 = cities_df["Lat"]
y_axis1 = cities_df["Max Temp"]
lattemp = plt.scatter(x_axis1, y_axis1, marker="o",facecolors="navy", 
                     edgecolors="black",s=20, alpha=0.75)
plt.grid()

# Incorporate the other graph properties
date = plotheaderdate.strftime("%x")
plt.title(f"City Latitude vs. Max Temperature ({date})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

# Save Figure
plt.savefig("JLT LatVSMaxTemp.png", bbox_inches="tight")
plt.show()
```
